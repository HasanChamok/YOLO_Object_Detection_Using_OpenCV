```python
import os
from glob import glob
import pandas as pd
from functools import reduce
from xml.etree import ElementTree as et
```

```python
#load xml files and load it in a list
xmlfiles = glob('./data_images/*.xml')  #*.xml will run all the xml file extention
xmlfiles
```

```python
#replace \\ with /
replace_text = lambda x: x.replace('\\','/')
xmlfiles = list(map(replace_text,xmlfiles))
xmlfiles
```

```python
#step-2: read xml files
#from each xml file we need to extract 
#filename, size(width,height), object(name,xmin,xmax,ymin,ymax)
tree = et.parse('./data_images/Cars0.xml')
root = tree.getroot()
#Now we can extract all the information from the xml file
print(root)
#extract filename [For single xml file]
image_name = root.find('filename').text
#width and height of the image
width = root.find('size').find('width').text
height = root.find('size').find('height').text
print("Image name : " + image_name + " Width : " + width + " Height : " + height)

#finding the object form the xml files
objs = root.findall('object')
print(objs)
obj = objs[0] 
name = obj.find('name').text
bndbox = obj.find('bndbox')
xmin = bndbox.find('xmin').text
xmax = bndbox.find('xmax').text
ymin = bndbox.find('ymin').text
ymax = bndbox.find('ymax').text
print([name , xmin , xmax , ymin , ymax])
```

```python
#For multiple xml file we have to run it into a loop
image_name = root.find('filename').text
#width and height of the image
width = root.find('size').find('width').text
height = root.find('size').find('height').text
objs = root.findall('object')
parser=[]
for obj in objs:
    name = obj.find('name').text
    bndbox = obj.find('bndbox')
    xmin = bndbox.find('xmin').text
    xmax = bndbox.find('xmax').text
    ymin = bndbox.find('ymin').text
    ymax = bndbox.find('ymax').text
    parser.append([image_name, width, height, name , xmin , xmax , ymin , ymax])

print(parser)
```

```python
#For multiple xml file we have to run it into a loop
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()
    image_name = root.find('filename').text
    #width and height of the image
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text
    objs = root.findall('object')
    parser=[]
    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name, width, height, name , xmin , xmax , ymin , ymax])

    return parser
```

```python
parser_all = list(map(extract_text,xmlfiles))
print(parser_all)
```
